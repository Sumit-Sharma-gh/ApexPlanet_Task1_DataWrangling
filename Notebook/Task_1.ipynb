```python
import pandas as pd
```

```python
df = pd.read_csv('Superstore.csv')
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

### The dataset contains 11 missing values in the Postal Code column, while all other fields are complete.

```python
df[df['Postal Code'].isnull()][['City','State','Region']]
```

```python
df.duplicated().sum()
```

```python
df['Postal Code'] = df['Postal Code'].fillna('Unknown')
```

```python
df['Postal Code'] = df['Postal Code'].astype(str)
```

```python
df.isnull().sum()
df['Postal Code'].value_counts().head()
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df[['Order Date','Ship Date','Shipping Days','Order Year','Order Month']].head()
```
